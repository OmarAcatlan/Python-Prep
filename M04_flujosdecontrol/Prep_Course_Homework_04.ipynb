1) Crear una variable que contenga un elemento del conjunto de números enteros y luego imprimir por pantalla si es mayor o menor a cero

```python
def es_primo(numero):
    if numero <= 1:
        return False
    for i in range(2, int(numero ** 0.5) + 1):
        if numero % i == 0:
            return False
    return True

# Pedir al usuario que ingrese un número
numero = int(input("Ingresa un número: "))

if es_primo(numero):
    print(numero, "es un número primo.")
else:
    print(numero, "no es un número primo.")
```

6) Utilizar un ciclo while para realizar el factorial de un número guardado en una variable, sólo si la variable contiene un número entero mayor a 0

```python
def factorial(num):
    if num == 0 or num == 1:
        return 1
    if num == 2:
        return 2
    resultado = 1
    i = 1
    while i <= num:
        resultado *= i
        i += 1
    return resultado
print(factorial(5))
```

11) En los puntos 9 y 10, se diseño un código que encuentra números primos y además se lo optimizó. ¿Es posible saber en qué medida se optimizó?

```python
def num_primo(numero):
    ciclos = 0
    if numero < 1:
        return False, ciclos
    if numero == 1 or numero == 2 or numero == 3:
        return True, ciclos
    if numero % 2 == 0:
        return False, ciclos
    i = 3
    while i <= int(numero**0.5) + 1:
        ciclos += 1
        if numero % i == 0:
            return False, ciclos
        i += 2
    return True, ciclos

print("Los números primos del 1 al 30 son:")
iteracion = 0
primos_encontrados = 0
for numero in range(1, 31000):
    es_primo, ciclos = num_primo(numero)
    iteracion += ciclos
    if es_primo:
        print(numero, ciclos)
        primos_encontrados += 1
    else:
        print(numero, "No es primo")
print("El total de iteraciones es:" + str(iteracion))
print("El total de números primos encontrados es: " + str(primos_encontrados))
```

13) Aplicando continue, armar un ciclo while que solo imprima los valores divisibles por 12, dentro del rango de números de 100 a 300

```python
n = 99
while(n <= 300):
    n += 1
    if (n % 12 != 0):
        continue
    print(n, ' es divisible por 12')
```

14) Utilizar la función **input()** que permite hacer ingresos por teclado, para encontrar números primos y dar la opción al usario de buscar el siguiente

```python
n = 1
sigue = 1
primo = True
while (sigue == 1):
    for div in range(2, n):
        if (n % div == 0):
            primo = False
            break
    if (primo):
        print(n)
        print('¿Desea encontrar el siguiente número primo?')
        if (input() != '1'):
            print('Se finaliza el proceso')
            break
    else:
        primo = True
    n += 1
```
